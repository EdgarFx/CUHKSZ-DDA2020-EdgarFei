```python
import re
import numpy as np
from sklearn import datasets
from sklearn.multiclass import OutputCodeClassifier
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.svm import SVC

class SVM(object):

    def __init__(self, training_dataset_, test_dataset_):
        self.training_dataset = training_dataset_
        self.test_dataset = test_dataset_
        self.classes = {}
        self.X_train = None
        self.Y_train = None
        self.X_test = None
        self.Y_test = None

        self.support_indecies = None
        self.train_errors = None
        self.val_errors = None
        self.train_loss = None
        self.val_loss = None
        
    def read_data(self):
        f = open(self.training_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        names, self.Y_train = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.X_train = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_train = np.append(self.X_train, np.array([np.array(row[:-1]).astype(float)]), axis = 0)
        f = open(self.test_dataset, 'r')
        f.close()
        
        f = open(self.test_dataset, 'r')
        rows = list(re.split(' ', row) for row in re.split('\n', f.read())[:-1])
        names, self.Y_test = np.unique(list(row[-1] for row in rows), return_inverse=True)
        self.X_test = np.empty((0,4), float)
        f.close()
        for row in rows:
            self.X_test = np.append(self.X_test, np.array([np.array(row[:-1]).astype(float)]), axis = 0)
        
    def SVM(self):
        
        global setosa_Y_train,versicolor_Y_train, virginica_Y_train
        
        # separate data into two class
        setosa_Y_train = []
        versicolor_Y_train = []
        virginica_Y_train = []
        setosa_Y_test = []
        versicolor_Y_test = []
        virginica_Y_test = []
        
        for i in range(len(self.Y_train)):
            if self.Y_train[i] == 0:
                setosa_Y_train.append(1)
                versicolor_Y_train.append(-1)
                virginica_Y_train.append(-1)
            elif self.Y_train[i] == 1:
                setosa_Y_train.append(-1)
                versicolor_Y_train.append(1)
                virginica_Y_train.append(-1)
            else:
                setosa_Y_train.append(-1)
                versicolor_Y_train.append(-1)
                virginica_Y_train.append(1)

        
        # fit svm model
        clf_setosa = SVC(kernel="linear",C=1e5)
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(kernel="linear",C=1e5)
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(kernel="linear",C=1e5)
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
        # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_
        
        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)
        
        
        # write file
        file = open("SVM_linearly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.coef_[0])):
            if i < len(clf_setosa.coef_[0])-1:
                file.write(str(clf_setosa.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(',')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.coef_[0])):
            if i < len(clf_versicolor.coef_[0])-1:
                file.write(str(clf_versicolor.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.coef_[0])):
            if i < len(clf_virginica.coef_[0])-1:
                file.write(str(clf_virginica.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        
        return
        
        
        
    def SVM_slack(self,C):
        
        # fit svm model
        clf_setosa = SVC(kernel="linear",C=C)
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(kernel="linear",C=C)
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(kernel="linear",C=C)
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
        # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_

        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)
        
        
        # write file
        file = open("SVM_slack_"+str(C)+"ly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.coef_[0])):
            if i < len(clf_setosa.coef_[0])-1:
                file.write(str(clf_setosa.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.coef_[0])):
            if i < len(clf_versicolor.coef_[0])-1:
                file.write(str(clf_versicolor.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.coef_[0])):
            if i < len(clf_virginica.coef_[0])-1:
                file.write(str(clf_virginica.coef_[0][i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.coef_[0][i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        
        
        return 
    
    def SVM_kernel_poly2(self,C):
        
        # fit svm model
        clf_setosa = SVC(C=C,kernel='poly',degree=2,gamma=1)
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(C=C,kernel='poly',degree=2,gamma=1)
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(C=C,kernel='poly',degree=2,gamma=1)
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
       # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_
        
        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)
        
        
        # write file
        file = open("SVM_poly2ly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        

        return 
    
    def SVM_kernel_poly3(self,C):
    
        # fit svm model
        clf_setosa = SVC(C=C,kernel='poly',degree=3,gamma=1)
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(C=C,kernel='poly',degree=3,gamma=1)
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(C=C,kernel='poly',degree=3,gamma=1)
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
         # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_
        
        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)
        
      
        # write file
        file = open("SVM_poly3ly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        

        return 
    
    
    def SVM_kernel_rbf(self,C):
    
        # fit svm model
        clf_setosa = SVC(C=C,kernel='rbf',gamma= 0.5)
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(C=C,kernel='rbf',gamma= 0.5)
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(C=C,kernel='rbf',gamma= 0.5)
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
         # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_
        
        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)

          
        # write file
        file = open("SVM_rbfly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        
        return 
    
    def SVM_kernel_sigmoid(self,C):
    
        # fit svm model
        clf_setosa = SVC(C=C,kernel='sigmoid')
        clf_setosa.fit(self.X_train,setosa_Y_train)
        
        clf_versicolor = SVC(C=C,kernel='sigmoid')
        clf_versicolor.fit(self.X_train,versicolor_Y_train)
        
        clf_virginica = SVC(C=C,kernel='sigmoid')
        clf_virginica.fit(self.X_train,virginica_Y_train)
        
        
         # find support vectors indices
        svi_setosa = clf_setosa.support_
        svi_versicolor = clf_versicolor.support_
        svi_virginica = clf_virginica.support_
        
        
        # find train error and test error
        predict_Y_train = []
        dec_setosa_train = clf_setosa.decision_function(self.X_train)
        dec_versicolor_train = clf_versicolor.decision_function(self.X_train)
        dec_virginica_train = clf_virginica.decision_function(self.X_train)
        for i in range(len(dec_setosa_train)):
            max_val = max(dec_setosa_train[i],dec_versicolor_train[i],dec_virginica_train[i])
            if max_val in dec_setosa_train:
                predict_Y_train.append(0)
            elif max_val in dec_versicolor_train:
                predict_Y_train.append(1)
            else:
                predict_Y_train.append(2)
        
        train_error = sum(self.Y_train != predict_Y_train)/len(self.Y_train)
        
        predict_Y_test = []
        dec_setosa_test = clf_setosa.decision_function(self.X_test)
        dec_versicolor_test = clf_versicolor.decision_function(self.X_test)
        dec_virginica_test = clf_virginica.decision_function(self.X_test)
        for i in range(len(dec_setosa_test)):
            max_val = max(dec_setosa_test[i],dec_versicolor_test[i],dec_virginica_test[i])
            if max_val in dec_setosa_test:
                predict_Y_test.append(0)
            elif max_val in dec_versicolor_test:
                predict_Y_test.append(1)
            else:
                predict_Y_test.append(2)
        
        test_error = sum(self.Y_test != predict_Y_test)/len(self.Y_test)
        
        
        # write file
        file = open("SVM_sigmoidly.txt",'w')
        file.write(str(train_error))
        file.write('\n')
        file.write(str(test_error))
        file.write('\n')
        for i in range(len(clf_setosa.intercept_)):
            if i < len(clf_setosa.intercept_)-1:
                file.write(str(clf_setosa.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_setosa.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_setosa)):
            if i < len(svi_setosa)-1:
                file.write(str(svi_setosa[i]))
                file.write(', ')
            else:
                file.write(str(svi_setosa[i]))
                file.write('\n')
        for i in range(len(clf_versicolor.intercept_)):
            if i < len(clf_versicolor.intercept_)-1:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_versicolor.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_versicolor)):
            if i < len(svi_versicolor)-1:
                file.write(str(svi_versicolor[i]))
                file.write(', ')
            else:
                file.write(str(svi_versicolor[i]))
                file.write('\n')
        for i in range(len(clf_virginica.intercept_)):
            if i < len(clf_virginica.intercept_)-1:
                file.write(str(clf_virginica.intercept_[i]))
                file.write(', ')
            else:
                file.write(str(clf_virginica.intercept_[i]))
                file.write('\n')
        for i in range(len(svi_virginica)):
            if i < len(svi_virginica)-1:
                file.write(str(svi_virginica[i]))
                file.write(', ')
            else:
                file.write(str(svi_virginica[i]))
                file.write('\n')
 
        file.close()
        

        return 
    
         
```
